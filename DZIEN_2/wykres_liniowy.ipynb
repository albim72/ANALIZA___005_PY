```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#przygotowanie danych
t = np.arange(0.0,2.0,0.01)
s = 1+np.sin(2*np.pi*t)
```

```python
#przygotowanie wykresu
fig,ax = plt.subplots()
ax.plot(t,s)
ax.set(xlabel='czas [s]',ylabel='napięcie prądu [mV]', title='pomiar napięcia prądu w czasie t')
ax.grid(True)
fig.savefig("wykres_sin.png")
plt.show()
```

```python
#diagram kołowy
labels = 'Warszawa','Kraków','Lublin','Katowice','Zakopane'
print(type(labels))
```

```python
sizes = [56,34,27,28,31]
explode = (0,0.1,0,0,0.2)
```

```python
fig,ax = plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%', shadow=True,startangle=90)
ax.axis('equal')
plt.show()
```

```python
#wykres słupkowy
marki = ['audi','toyota','mercedes','honda','kia','jeep','seat']
sprzed = [12.7,28.9,9.8,16.8,23.5,7.8,28.3]

bar_labels = marki[:]
bar_colors = ['tab:red','tab:green','tab:blue','tab:orange','#FFFF00','#AB4519','#19FF03']
```

```python
fig,ax = plt.subplots()
ax.bar(marki,sprzed,label=bar_labels,color=bar_colors)
ax.set_xlabel('marki samochodów')
ax.set_ylabel('sprzedaż samochodów [tys]')

ax.legend(title='marki')
plt.show()
```
