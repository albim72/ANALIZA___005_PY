```python
import pandas as pd
```

```python
unames = ['user_id','gender','age','occupation','zip']
users = pd.read_table('dane/users.dat',sep="::",engine='python',header=None,names=unames)

rnames = ['user_id','movie_id','rating','timestamp']
ratings = pd.read_table('dane/ratings.dat',sep="::",engine='python',header=None,names=rnames)

mnames = ['movie_id','title','genres']
movies = pd.read_table('dane/movies.dat',sep="::",engine='python',encoding='ISO-8859-1',header=None,names=mnames)
```

```python
users[:5]
```

```python
movies[:5]
```

```python
ratings[:5]
```

```python
data = pd.merge(pd.merge(ratings,users),movies)
data
```

```python
mean_ratings = data.pivot_table('rating',index='title',columns='gender',aggfunc='mean')
mean_ratings[:5]
```

```python
#ile ocen uzyskał każdy z filmów
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title>=250]
active_titles[:10]
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings[:10]
```

```python
top_female_ratings = mean_ratings.sort_values(by='F',ascending=False)
top_female_ratings[:10]
```

```python
mean_ratings["diff"] = mean_ratings['M'] - mean_ratings['F']
```

```python
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:10]
```
