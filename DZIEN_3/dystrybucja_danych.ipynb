```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

wygeneruj tablicę 1D(wektor) zawierającą 100 wartości, gdzie każda wartość musi być jedną z: [3,5,8,11].<br>
Prawdopodobieństwo wystąpień dla poszczególnych wartości:<br>
3 -> 0,1<br>
5 -> 0,25<br>
8 -> 0,6<br>
11 -> 0,05<br>

```python
x = random.choice([3,5,8,11],p=[0.1,0.25,0.6,0.05],size=(100))
print(x)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)

plt.title("Rozrzut Rachunek - Napiwek")
plt.xlabel("całkowity rachunek")
plt.ylabel("napiwek")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Rachunek a rodzaj dnia")
plt.xlabel("Dzień")
plt.ylabel("całkowity rachunek")
plt.show()
```

```python
sns.kdeplot(tips['total_bill'],fill=True)

plt.title("Wykres gęstości dla rachunku calkowitego")
plt.xlabel("całkowity rachunek")
plt.ylabel("Gęstość")
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```
